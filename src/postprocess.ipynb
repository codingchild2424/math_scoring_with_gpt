```python
import pandas as pd
import os

results_path = "/workspace/company/math_scoring_with_gpt/results"
post_results_path = "/workspace/company/math_scoring_with_gpt/post_results"
```

# 1-1

```python
number_1_1_path = os.path.join(results_path, "number_1_1_eval.tsv")

print(number_1_1_path)
```

```python
df_1_1 = pd.read_csv(number_1_1_path, sep="\t")

df_1_1.head()
```

```python
# \n\n -> \n
df_1_1['bot_response'] = df_1_1['bot_response'].apply(lambda x: x.replace("\n\n", "\n"))

df_1_1.head()
```

```python
# df_1_1['bot_response'] split by \n
df_1_1['bot_response_split'] = df_1_1['bot_response'].apply(lambda x: x.split("\n"))
```

```python
df_1_1.head()
```

```python
# make new two columns using bot_response_split column, which is a list
df_1_1['score'] = df_1_1['bot_response_split'].apply(lambda x: x[0])
df_1_1['reason'] = df_1_1['bot_response_split'].apply(lambda x: x[1])

df_1_1.head()
```

```python
# preprocessing in score column, erase "점수: "
df_1_1['score'] = df_1_1['score'].apply(lambda x: x.replace("점수: ", ""))

df_1_1.head()
```

```python
# preprocessing in reason column, erase "이유: "
df_1_1['reason'] = df_1_1['reason'].apply(lambda x: x.replace("이유: ", ""))
```

```python
df_1_1.head()
```

```python
df_1_1_pre = df_1_1[['user_id', 'user_answer', 'score', 'reason']]

df_1_1_pre.head()
```

```python
df_1_1_pre_path = os.path.join(post_results_path, "number_1_1_eval_pre.tsv")

df_1_1_pre.to_csv(df_1_1_pre_path, sep="\t", index=False, encoding='utf-8')
```
