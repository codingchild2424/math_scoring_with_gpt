```python
import pandas as pd

# Data Path
data_path = "/workspace/company/math_scoring_with_gpt/data/number_2.tsv"

# Load Data
df = pd.read_csv(data_path, sep="\t")
```

# 2

```python
df.head()
```

```python
answer1 = []
answer2 = []
answer3 = []

for row in df.iterrows():

    answer1_list = []
    answer2_list = []
    answer3_list = []

    for i in str(row[1]['답안']).split("\n"):
        ##print("row: ", row[1]['user_id'])
        answer1_list.append(str(i).split(",")[0])
        answer2_list.append(str(i).split(",")[1])
        answer3_list.append(str(i).split(",")[2])

    answer1.append(", ".join(answer1_list).replace("\n", ""))
    answer2.append(", ".join(answer2_list).replace("\n", ""))
    answer3.append(", ".join(answer3_list).replace("\n", ""))

print(len(answer1))
print(len(answer2))
print(len(answer3))

df['답안1'] = answer1
df['답안2'] = answer2
df['답안3'] = answer3

df.head()
    
```

```python
# save result
save_path = "/workspace/company/math_scoring_with_gpt/data/number_2.tsv"
df.to_csv(save_path, sep="\t", index=False)
```
