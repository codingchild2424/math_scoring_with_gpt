```python
import pandas as pd

number1_path = "/workspace/company/math_scoring_with_gpt/data/number_1.xlsx"

number1_df = pd.read_excel(number1_path, sheet_name="Sheet1")

number1_df.head()
```

```python
number1_df.shape
```

```python
number1_df = number1_df[['user_id', '답안', '1. 사다리꼴이 사각형임을 알고 있는가?', '2. 평행의 개념을 알고 있는가?']]
number1_df.head()
```

```python
number1_df.shape
```

```python
# erase '\t' in number1_df's '답안' column 
number1_df['답안'] = number1_df['답안'].str.replace('\t', ' ')
```

```python
# save number1_df as tsv file

number1_df.to_csv("/workspace/company/math_scoring_with_gpt/data/number_1.tsv", sep='\t', index=False)
```

```python
# load number1_df as tsv file

number1_df_test = pd.read_csv("/workspace/company/math_scoring_with_gpt/data/number_1.tsv", sep='\t')

number1_df_test.head()
```
